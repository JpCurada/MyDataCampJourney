# 3. Additional Plot Types

Categorical Data
- Data which takes on a limited and fixed number of values
- Normally combined with numeric data
- Examples include:
    - Geography (country, state, region)
    - Gender
    - Ethnicity
    - Blood type
    - Eye color

Addn't Plot Types
1. Stripplot
- Displaying distribution of categorical data with individual points.
- Detecting outliers.

2. Swarmplot
- Displaying distribution of categorical data with non-overlapping points.
- Visualizing the density of data points.

3. Boxplot
- Summarizing distribution of numerical data.
- Identifying median, quartiles, and potential outliers.
- Comparing distributions between different categories.

4. Violinplot
- Displaying distribution and density of numerical data.
- Comparing distributions between different categories.
- Identifying modes and skewness in the data.

5. Boxenplot
- Providing detailed summary of the distribution for large datasets.
- Identifying quartiles, outliers, and other quantiles.
- Comparing distributions between different categories.

6. Barplot
- Comparing values or counts across different categories.
- Showing relationships between categorical variables and numerical values.
- Visualizing frequency or proportion of categories.

7. Pointplot
- Displaying mean or median values for different categories.
- Showing relationships between two categorical variables and their corresponding values.
- Comparing trends or patterns across categories.

8. Countplot
- Displaying frequency or count of each category in a categorical variable.
- Understanding the distribution and proportion of categories.
- Identifying imbalances or imputed values in the data.


```python
df = pd.read_csv('datasets/schoolimprovement2010grants.csv')
# Create the stripplot
sns.stripplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         jitter=True)

plt.show()
```

```python
# Create and display a swarmplot with hue set to the Region
sns.swarmplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         hue='Region')

plt.show()
```

```python
# Create a boxplot
sns.boxplot(data=df,
         x='Award_Amount',
         y='Model Selected')

plt.show()
plt.clf()
```

```python
# Create a violinplot with the husl palette
sns.violinplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='husl')

plt.show()
plt.clf()
```

```python
# Create a boxenplot with the Paired palette and the Region column as the hue
sns.boxenplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='Paired',
         hue='Region')

plt.show()
plt.clf()
```

```python
# Show a countplot with the number of models used with each region a different color
sns.countplot(data=df,
         y="Model Selected",
         hue="Region")

plt.show()
plt.clf()
```

```python
# Create a pointplot and include the capsize in order to show caps on the error bars
sns.pointplot(data=df,
         x='Model Selected',
         y='Award_Amount',
         capsize=.1)

plt.show()
plt.clf()
```

```python
# Create a barplot with each Region shown as a different color
sns.barplot(data=df,
         y='Award_Amount',
         x='Model Selected',
         hue='Region')

plt.show()
plt.clf()
```

The residual plot can interpret the second order polynomial and plot the residual values. In this example, the values are more randomly distributed, so a second order equation is likely more appropriate for this data.



**Categorical Variables**

Seaborn also supports regression plots with categorical variables. It might be interesting to see how rentals change over the various months. In this example, using the jitter parameter makes it easier to see the individual distribution of the rental values for each month.

```python
sns.regplot(data=df, x='mnth', y='total_rentals', x_jitter=.1 ,order=2)
```

**Estimators**

- An x_estimator can be useful for highlighting trends

In some cases, even with the jitter, it might be difficult to see if there are any trends based on the value of the variable. Using an estimator for the x value can provide another helpful view of the data. This simplified view shows a trend consistent with the seasons in Washington, DC.

```python
import numpy as np
sns.regplot(data=df, x='mnth', y='total_rentals', x_estimator=np.mean ,order=2)
```

**Binning the data**

- x_bins can be used to divide the data into discrete bins
- The regression line is still fit against all the data

When there are continuous variables, it can be helpful to break them into different bins. In this case, we can divide the temperature into four bins and Seaborn will take care of calculating those bins for us and plotting the results. This is much quicker than trying to use pandas or some other mechanism to create the bins. 


```python
sns.regplot(data=df, x='mnth', y='total_rentals', x_bins=4)
```

# 4. Creating Plots on Data Aware Grids

**FacetGrid**
- The FacetGrid is foundational for many data aware grids
- It allows the user to control how data is distributed across columns, rows and hue
- Once a Facetrid is created, the plot type must be mapped to the grid

**catplot()**
- The catplot is a simpler a way to use a FacetGrid for categorical data
- Combines the facetting and mapping process into 1 function

```python
df = pd.read_csv('datasets/college_datav3.csv')
g = sns.FacetGrid (df, col='HIGHDEG')
g.map(sns.boxplot, 'Tuition', order=['1', '2', '3', '4'])
```

```python
sns.catplot(x="Tuition", data=df, col="HIGHDEG", kind='box')
```

**FacetGrid for regression**
- FacetGrid() can also be used for scatter or regression plots.

```python
g = sns.FacetGrid (df, col='HIGHDEG')
g.map(plt.scatter, 'Tuition', 'SAT_AVG_ALL')
```

**Implot**
- lmplot plots scatter and regression plots on a FacetGrid

```python
sns.lmplot(data=df, x='Tuition', y='SAT_AVG_ALL', col="HIGHDEG", fit_reg=False)
```

**Using JointGrid and jointplot**

A JointGrid() allows us to compare the distribution of data between two variables. A JointGrid() makes use of scatter plots, regression lines, as well as histograms, distribution plots, and kernel density estimates to give us insights into our data.

```python
df = pd.read_csv('datasets/college_datav3.csv')
g = sns.JointGrid(data=df, x='Tuition', y='ADM_RATE_ALL')
g.plot(sns.regplot, sns.histplot)
```

```python
# Advanced JointGrid
g = sns.JointGrid(data=df, x='Tuition', y='ADM_RATE_ALL')
g = g.plot_joint(sns.kdeplot)
g = g.plot_marginals(sns.kdeplot, shade=True)
```

```python
sns.jointplot(data=df, x='Tuition', y='ADM_RATE_ALL', kind='hex')
```

```python
# Customizing a jointplot
g = (sns.jointplot(x='Tuition',
                  y='ADM_RATE_ALL',
                  kind='scatter',
                  xlim=(0,25000),
                  data=df.query('UG < 2500 & Ownership == "Public"')).plot_joint(sns.kdeplot))
```
