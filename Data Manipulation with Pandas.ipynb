### 1.2. Sorting and Subsetting

```python
nf_df.sort_values("release_year").head()
```

```python
# Subsetting columns
ph_tv_shows_df = nf_df[nf_df['country'] == 'Philippines']
ph_tv_shows_df.sort_values('duration')
```

```python
# Subsetting Columns 
ph_movie_df = nf_df[(nf_df['type'] == 'Movie') & (nf_df['country'] == 'Philippines')]
ph_movie_df.sort_values('release_year')
```

```python
# Subsetting Columns 
skdrama_df = nf_df[(nf_df['type'] == 'TV Show') & (nf_df['country'] == 'South Korea')]
skdrama_df.sort_values('release_year')
```

```python
# Subsetting Rows
cathy_garcia_films_df = ph_movie_df[ph_movie_df['director'].isin(['Cathy Garcia-Molina'])]
cathy_garcia_films_df
```
